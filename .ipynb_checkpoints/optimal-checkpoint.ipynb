```python
###########################
# Project 6
# ESE 572
# Optimal Routing
###########################

import numpy as np
import math
import networkx as nx
import matplotlib.pyplot as plt
from heapq import heappush, heappop
```

```python
# For the network given below, the goal of this project is to develop a routing assignment that 
# optimizes the delay in the network.

# Problem Assumptions
#     1. For each link, the delay is assumed to be M/M/1
#     2. All traffic flows in the network, other than the one from node A to node B, have already been routed.  
#        Your problem is to determine the routing assignment for the remaining traffic flow from node A to node B.
#     3. The numbers shown in the network (without parenthesis) are the average delays for individual links 
#        before the traffic from node A to node B is routed
#     4. The numbers shown inside the parentheses are the link capacities.


# First, identify three best candidate routes from node A to node B.  This is not meant to be based on any 
# mathematical reasons – just pick three routes that your group believes to be reasonable and explain why these 
# routes have been chosen.  Then, distribute the 5 units of flow from node A to node B, shown in the figure as 
# r = 5, to among the three candidate routes, which optimize the overall delay in the network
```

```python
########################################
# Step 1: Three Best Candidates
# Dijkstra’s Algorithm 
# Undirected and weighted graph
########################################

class Graph:
    def __init__(self,nodes):
        self.nodes = nodes
        self.graph = []
        
    def add_e(self,source,dest,weight):
        self.graph.append([source,dest,weight])
    
    def dijkstra(self, source, end):
        pos = {0: (10, 20), 1: (15, 27.5), 2: (15, 22.5), 3: (15, 17.5), 4: (15, 12.5),
               5: (25, 27.5), 6: (25, 22.5), 7: (25, 17.5), 8: (25, 12.5), 9: (30, 20)}
        dist = [float('inf')] * self.nodes # Initialize distances to all nodes to infinity
        dist[source] = 0  # Set distance to source node to 0
        visited = [False] * self.nodes # Create an array to keep track of visited nodes
        path = [[] for _ in range(self.nodes)] # Create an array to keep track of the shortest path to each node
        pq = [(0, source)] # Create a priority queue to store nodes to visit
        it_num = 1
        while pq: # Loop through the priority queue until it's empty
            d, u = heappop(pq) # Pop the node with the smallest distance from the priority queue
            if visited[u]: # If the node has already been visited, skip it
                continue
            visited[u] = True # Mark the node as visited
            
            for edge in self.graph: # Update the distances to all neighboring nodes
                if edge[0] == u or edge[1] == u: # Check if the edge starts or ends at the current node
                    v = edge[1] if edge[0] == u else edge[0] # Get the other node in the edge
                    w = edge[2] # Get the weight of the edge
                    if dist[u] + w < dist[v]:
                        dist[v] = dist[u] + w
                        path[v] = path[u] + [u] # Add the neighboring node to the priority queue
                        heappush(pq, (dist[v], v)) # If we've reached the end node, we can stop searching
            if visited[end]:
                break

        full_path = path[end] + [end]
        drawn_edges = []
        for fp in range(len(full_path)-1):
            s_node = full_path[fp]
            d_node = full_path[fp+1]
            for lst in self.graph:
                if (lst[0]==s_node and lst[1]==d_node) or (lst[1]==s_node and lst[0]==d_node):
                    path_w = lst[2]
                    break
            drawn_edges.append([s_node,d_node,path_w])

            # Plotting iterations
            G_visual = nx.Graph()
            iteration_cost = 0
            for vn in range(len(visited)):
                G_visual.add_node(vn) # draw all nodes
            for each in drawn_edges:
                G_visual.add_edge(each[0], each[1], weight=each[2])
                iteration_cost += each[2]

            labels = nx.get_edge_attributes(G_visual,'weight')
            plt.figure()
            nx.draw_networkx(G_visual, pos)
            nx.draw_networkx_edge_labels(self,pos,edge_labels=labels)
            plt.title('Iteration %d, Source: %d, Dest: %d, Iteration Cost: %d' % 
                      (it_num, source, end, iteration_cost), fontsize=12)
            it_num += 1
            
        return dist[end], path[end] + [end] # Return the shortest distance and the shortest path to the end node
```

```python
########################################
# First Best Candidate Route
########################################

G1 = Graph(10)
G1.add_e(0,1,2)
G1.add_e(0,3,1)
G1.add_e(0,4,2)
G1.add_e(1,2,1)
G1.add_e(2,3,1)
G1.add_e(3,4,4)
G1.add_e(1,5,6)
G1.add_e(2,6,3)
G1.add_e(3,7,2)
G1.add_e(4,8,1)
G1.add_e(5,6,1)
G1.add_e(6,7,3)
G1.add_e(7,8,4)
G1.add_e(5,9,1)
G1.add_e(6,9,3)
G1.add_e(8,9,2)

print('A to B')
dist_1, path_1 = G1.dijkstra(0, 9)
print(f'Shortest path distance: {dist_1}')
print('Shortest path:', path_1)
print()
```

```python
########################################
# Second Best Candidate Route
########################################

G2 = Graph(10)
G2.add_e(0,1,2)
G2.add_e(0,3,1)
G2.add_e(0,4,8) # Increase delay cost on this link, since lowest capacity, to find second best route
G2.add_e(1,2,1)
G2.add_e(2,3,1)
G2.add_e(3,4,4)
G2.add_e(1,5,6)
G2.add_e(2,6,3)
G2.add_e(3,7,2)
G2.add_e(4,8,1)
G2.add_e(5,6,1)
G2.add_e(6,7,3)
G2.add_e(7,8,4)
G2.add_e(5,9,1)
G2.add_e(6,9,3)
G2.add_e(8,9,2)

print('A to B')
dist_2, path_2 = G2.dijkstra(0, 9)
print(f'Shortest path distance: {dist_2}')
print('Shortest path:', path_2)
print()
```

```python
########################################
# Third Best Candidate Route
########################################

G3 = Graph(10)
G3.add_e(0,1,2)
G3.add_e(0,3,1)
G3.add_e(0,4,8) # Increase delay cost on this link, since lowest capacity, to find second best route
G3.add_e(1,2,1)
G3.add_e(2,3,1)
G3.add_e(3,4,4)
G3.add_e(1,5,6)
G3.add_e(2,6,9) # Increase delay cost on this link, since lowest capacity, to find third best route
G3.add_e(3,7,2)
G3.add_e(4,8,1)
G3.add_e(5,6,1)
G3.add_e(6,7,3)
G3.add_e(7,8,4)
G3.add_e(5,9,1)
G3.add_e(6,9,3)
G3.add_e(8,9,2)

print('A to B')
dist_3, path_3 = G3.dijkstra(0, 9)
print(f'Shortest path distance: {dist_3}')
print('Shortest path:', path_3)
print()
```

```python
########################################
# Step 2: Distribute Units of Flow
# Using 3 route candidates & capacities
########################################

r = 5
print('Flow to distribute from node A to node B')
print('r =',r)
print('----------------------------------------------')

# dictionary of all link capacities (in parenthesis on graph picture)
capacities = {(0,1): 4,(0,3): 5,(0,4): 2,(1,2): 1,(2,3): 3,(3,4): 8,(1,5): 4,(2,6): 3,(3,7): 2,(4,8): 3,(5,6): 4,
              (6,7): 2,(7,8): 2,(5,9): 3,(6,9): 2,(8,9): 4,
              # making undirected by reverse tuples for each link
              (1,0): 4,(3,0): 5,(4,0): 2,(2,1): 1,(3,2): 3,(4,3): 8,(5,1): 4,(6,2): 3,(7,3): 2,(8,4): 3,(6,5): 4,
              (7,6): 2,(8,7): 2,(9,5): 3,(9,6): 2,(9,8): 4}

# paths and total delay cost for 3 candidate routes
paths = [path_1, path_2, path_3]
dists = [dist_1, dist_2, dist_3]

used_paths = []
distributions = []
# ditributing the flow by iterating though candidates in order of smallest total link delay until r = 0
while(paths and r > 0):
    # choose candidate with smallest total delay
    min_ind = dists.index(min(dists))
    min_path = paths[min_ind]
    min_dist = dists[min_ind]
    
    # remove candidate from list of unused candidates
    paths.remove(paths[min_ind])
    dists.remove(dists[min_ind])

    print('Min Path:', min_path)
    print('Min Dist:', min_dist)

    tups = []
    caps = []
    # make tuples for each link in route & list of their link capacities
    for p in range(len(min_path)-1):
        tup = (min_path[p],min_path[p+1])
        tups.append(tup)
        caps.append(capacities[tup])

    print('Links for path:', tups)
    print('Link capacities for path:', caps)

    # choose smallest capacitity/max flow through that path
    max_flow = min(caps)
    print('Flow for Path:',max_flow)

    # distribute max flow units through this link, decreasing r (flow still needing distributed)
    r = r - max_flow
    print('r =', r)
    print('----------------------------------------------')
    used_paths.append(min_path)
    distributions.append(max_flow)

print('Paths used to distribute from A to B:', used_paths)
print('Distributions:', distributions)
```

```python
########################################
# Step 3: Visualizing Optimal Solution
# Used candidate routes & how much was
# distributed to each from r = 5
########################################

G_final = nx.Graph()
pos = {0: (10, 20), 1: (15, 27.5), 2: (15, 22.5), 3: (15, 17.5), 4: (15, 12.5),
               5: (25, 27.5), 6: (25, 22.5), 7: (25, 17.5), 8: (25, 12.5), 9: (30, 20)}

for nod in range(10):
    G_final.add_node(nod) # draw all nodes

d_edges = []
for ip in range(len(used_paths)):
    used_path = used_paths[ip]
    used_flow = distributions[ip]
    for up in range(len(used_path)-1):
        s_node = used_path[up]
        d_node = used_path[up+1]
        d_edges.append([s_node,d_node,used_flow])

for each in d_edges:
    # coloring different routes
    if each[2] == distributions[0]:
        color_val = 'r'
    elif each[2] == distributions[1]:
        color_val = 'k'
    else:
        color_val = 'g'
    G_final.add_edge(each[0], each[1], color=color_val, weight=each[2])

labels = nx.get_edge_attributes(G_final,'weight')
colors = nx.get_edge_attributes(G_final,'color').values()

# plot the graph
plt.figure()
nx.draw_networkx(G_final, pos, edge_color=colors)
nx.draw_networkx_edge_labels(G_final,pos,edge_labels=labels)
plt.title('Optimal Distribution of r = 5 flow from node A to node B', fontsize=12)
print('Edge labels represent flow much flow is distributed to that link.')
```
